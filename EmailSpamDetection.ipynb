```python
email_text = email_data['text']
```

```python
y = email_data['label_num'].to_numpy()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
countVec = CountVectorizer()
X_data = countVec.fit_transform(email_text)
```

```python
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X_data, y)
```

```python
def prob_class(y):
    c1_arr = np.where(y == 1)
    c1 = len(c1_arr[0])
    pc1 = c1/len(y)
    pc0 = 1-pc1
    return [pc0, pc1]
```

```python
pc = prob_class(Y_train)
```

```python
X_word_tr = X_train.T
```

```python
M,N = X_train.shape
pc_mat = np.zeros((2,N))
s_count,ns_count = (0,0)

for y in range(len(Y_train)):
    if Y_train[y] == 1:
        old_s_count = s_count
        s_count += 1
        pc_mat[1,:] = (old_s_count * pc_mat[1,:] + X_train[y])/s_count

    if Y_train[y] == 0:
        old_ns_count = ns_count
        ns_count += 1
        pc_mat[0,:] = (old_ns_count * pc_mat[0,:] +X_train[y])/ns_count
```

```python
def multiNB(X):
    y_pred = []
    X = X.toarray()
    print(X)
    for email in X:
        eml = np.where(email > 0)
        n = len(eml[0])
        log_term = math.log(math.factorial(n)) - sum(math.log(math.factorial(xi)) for xi in email)
        log_prob_sum0 = sum(xi * math.log(pi) for xi, pi in zip(email, pc_mat[0,:]) if pi > 0)
        log_prob_sum1 = sum(xi * math.log(pi) for xi, pi in zip(email, pc_mat[1,:]) if pi > 0)
        lp0 = log_term + log_prob_sum0 + math.log(pc[0])
        lp1 = log_term + log_prob_sum1 + math.log(pc[1])
        y_pred.append(1 if lp0 < lp1 else 0)
    return y_pred
```

```python
y_pred = multiNB(X_test)
```

```python
accuracy = sklearn.metrics.accuracy_score(Y_test, y_pred)
print(f'Accuracy of Custom Multinomial NB: {accuracy}')
```
